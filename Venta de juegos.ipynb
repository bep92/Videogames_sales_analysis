```python
#MOST GENRE SOLD BY YEAR
g = juegos.groupby(['Genre', 'Year']).count().reset_index()
g = g[['Genre', 'Year', 'Global_Sales']]
g_pivot = g.pivot(index ='Year'  , columns='Genre', values='Global_Sales')
figsize=(20,20)

g_pivot[['Action', 'Sports', 'Shooter', 'Role-Playing', 'Platform', 'Misc', 'Racing', 'Fighting', 'Simulation', 'Puzzle', 'Adventure', 'Strategy']].plot(
    color=['blue', 'red', 'yellow', 'orange', 'black', 'brown', 'pink', 'green', 'gold', 'silver', 'grey', 'magenta'], marker= 'o'
)

plt.title
plt.xlabel('año')
plt.ylabel('crecimiento')
plt.legend(title='genero')
plt.grid(True)
plt.show()
```

```python
#NA sales per year

NA_byear = juegos.groupby('Year')['NA_Sales'].sum()

plt.figure(figsize=(5,5))
plt.plot(NA_byear, marker = 'o', color='black', linestyle = '-')
plt.title('NA sales per year')
plt.xlabel('Año')
plt.ylabel('Crecimiento')
plt.grid(True)
plt.show()

```

```python
#EU sales per year
EU_byear = juegos.groupby('Year')['EU_Sales'].sum()

plt.figure(figsize=(5,5))
plt.plot(EU_byear, marker = 'o', color='Green', linestyle = '-')
plt.title('EU sales per year')
plt.xlabel('Año')
plt.ylabel('Crecimiento')
plt.grid(True)
plt.show()
```

```python
#JP sales per year
JP_byear = juegos.groupby('Year')['JP_Sales'].sum()

plt.figure(figsize=(5,5))
plt.plot(JP_byear, marker = 'o', color='Red', linestyle = '-')
plt.title('JP sales per year')
plt.xlabel('Año')
plt.ylabel('Crecimiento')
plt.grid(True)
plt.show()
```

```python
#Other sales per year
Ot_byear = juegos.groupby('Year')['Other_Sales'].sum()

plt.figure(figsize=(5,5))
plt.plot(Ot_byear, marker = 'o', color='Blue', linestyle = '-')
plt.title('Other sales per year')
plt.xlabel('Año')
plt.ylabel('Crecimiento')
plt.grid(True)
plt.show()
```

```python
#Global sales per year Global_Sales
GLO_byear = juegos.groupby('Year')['Global_Sales'].sum()

plt.figure(figsize=(5,5))
plt.plot(GLO_byear, marker = 'o', color='Gold', linestyle = '-')
plt.title('Global sales per year')
plt.xlabel('Año')
plt.ylabel('Crecimiento')
plt.grid(True)
plt.show()
```

We can observe that each country contains a similar graph due to the exponential growth of advertising over the years and also the evolution of people's taste for each genre, i.e. in Japan it is unlikely that the same sales scales for a certain type of genre will be repeated as in the United States or Europe.

```python
#COMPARATION EACH AREA IN EACH GENRE
```

```python
GroupGenre = juegos.groupby(['Genre']).sum().loc[:, 'NA_Sales':'Global_Sales']
GroupGenre['NA_Sales%'] = GroupGenre['NA_Sales']/GroupGenre['Global_Sales']
GroupGenre['EU_Sales%'] = GroupGenre['EU_Sales']/GroupGenre['Global_Sales']
GroupGenre['JP_Sales%'] = GroupGenre['JP_Sales']/GroupGenre['Global_Sales']
GroupGenre['Other_Sales%'] = GroupGenre['Other_Sales']/GroupGenre['Global_Sales']
plt.figure(figsize=(10,10))
sns.set(font_scale=0.7)
plt.subplot(211)
sns.heatmap(GroupGenre.loc[:, 'NA_Sales':'Other_Sales'], annot=True, fmt = '.1f')
plt.title('comparation of each area in genre')
plt.subplot(212)
sns.heatmap(GroupGenre.loc[:,'NA_Sales%':'Other_Sales%'], vmax=1, vmin=0, annot=True, fmt= '.2%')
plt.title('comparation of each area in each genre(percentage)')
plt.show()
```

As a conclusion, we can make it clear that the evolution of technology focused on advertising, was what allowed to increase the sales of video games and also the same advertising was the one that in turn was responsible for starting to include more people who previously found no interest in video games. Basically today, the company that spends more on advertising has a much higher percentage of obtaining higher sales, but the same will depend in turn on the type of genre of that game and in what type of country and / or continent is invested more in advertising. Since the public to which the game is focused always varies and that also allows it to attract more people who want to try new experiences in this virtual world that every day grows a little more and becomes more accepted with the passage of time.
